```python
#splitting the dataset
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
df_test_train, df_test_test=train_test_split(df_test, test_size=0.2,stratify=df_test['label'],random_state=1000)
features=df_test_train.drop(columns=['label']).columns
```

```python
#checking number rows in the train dataset
df_test_train['label'].value_counts()
```

```python
#checking number rows in the test dataset
df_test_test['label'].value_counts()
```

```python
#applying random oversampling-for balancing the dataset
msk=df_test_train['label']==1
num_to_oversample=len(df_test_train)-2*msk.sum()
df_test_positive_oversample=df_test_train[msk].sample(n=num_to_oversample,replace=True,random_state=1000)
df_test_train_oversample=pd.concat([df_test_train,df_test_positive_oversample])
df_test_train_oversample['label'].value_counts()
```

```python
#applying logistic regression to the balanced dataset(random oversampled)
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=1000)
lr.fit(df_test_train_oversample[features],df_test_train_oversample['label'])
y_predict=lr.predict_proba(df_test_test[features])[:,1]

from sklearn.metrics import roc_auc_score
roc_auc_score(df_test_test['label'],y_predict)
```

```python
#applying SMOTE
from  imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=1000)
x_newsample, y_newsample=smote.fit_resample(df_test_train[features],df_test_train['label'])
y_newsample.value_counts()
```

```python
#applying Logistic regression to SMOTE new sampled date
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=7000)
lr.fit(x_newsample,y_newsample)
y_predict=lr.predict_proba(df_test_test[features])[:,1]

from sklearn.metrics import roc_auc_score
roc_auc_score(df_test_test['label'],y_predict)
```
